### Data Cleaning

```python
cleaned_df = df.copy()
```

```python
# Starting with Invoice column. (Filter out invoices with C and A ) 
cleaned_df["Invoice"] = cleaned_df["Invoice"].astype("str") # (again this is converted into a string so regex can be used since we have to filter out specific invoice patters)

mask = (
    cleaned_df["Invoice"].str.match("^\\d{6}$") == True
)

cleaned_df = cleaned_df[mask] # cleaned_df[mask]  selects all rows from cleaned_df where the corresponding value in mask is True and filters out the rest

cleaned_df
```

```python
# StockCode (| means or)
cleaned_df["StockCode"] = cleaned_df["StockCode"].astype("str")

mask = (
    (cleaned_df["StockCode"].str.match("^\\d{5}$") == True)
    | (cleaned_df["StockCode"].str.match("^\\d{5}[a-zA-Z]+$") == True) # 5 digits followed by a letterS
    | (cleaned_df["StockCode"].str.match("^PADS$") == True)
)

cleaned_df = cleaned_df[mask]

cleaned_df
```

```python
# Missing customer ID
cleaned_df.dropna(subset=["Customer ID"], inplace=True)
# dropna - drop nulls,subset specifies a subset the column to perform the drop null function, 
# inplace specifies this operation to happen on the current dataframe so we dont have to reasign it.
```

### Feature engineering

```python
cleaned_df["SalesLineTotal"] = cleaned_df["Quantity"] * cleaned_df["Price"] 

cleaned_df
```

```python
# Aggregation means combining or grouping data in some way to derive summary metrics that help make sense of large or detailed data sets. 
# The groupby() function in pandas is used to group rows of a DataFrame by one or more columns and then apply an aggregation function to the grouped data.
# as_index is to prevent customer if from being the index

aggregated_df = cleaned_df.groupby(by = "Customer ID", as_index = False) \
.agg(
     MonetaryValue=("SalesLineTotal", "sum"),
        Frequency=("Invoice", "nunique"),
        LastInvoiceDate=("InvoiceDate", "max")
    )

aggregated_df.head(5)
```

```python
# To find out recency, how many days has it been since the last purchase

max_invoice_date = aggregated_df["LastInvoiceDate"].max() # To find the last purchase date in the dataset (since the data is very old)

aggregated_df["Recency"] = (max_invoice_date - aggregated_df["LastInvoiceDate"]).dt.days # how long has it been since the persons last transaction relative to the max invoice date in our data set

aggregated_df.head(5)
```

### KMeans Clustering

```python
max_k = 12

inertia = []
silhoutte_scores = []
k_values = range(2, max_k + 1) # As defined earlier, max_k is set to 12, so this range will be from 2 to 12 (inclusive).

for k in k_values:

    kmeans = KMeans(n_clusters=k, random_state=42, max_iter=1000)
    
#n_clusters=k: The number of clusters (k) to divide the data into.
#random_state=42: This ensures that the K-means algorithm's random initialization of cluster centroids is the same every time you run the code (so the results are reproducible).
#max_iter=1000: The maximum number of iterations K-means will run before stopping. This ensures it has enough time to converge.

    cluster_labels = kmeans.fit_predict(scaled_data_df)
    
#This fits the K-means model to the data (scaled_data_df), meaning it tries to find the best clustering solution.
#It then assigns each data point in the dataset to one of the k clusters and returns a list of cluster labels that tells  which cluster each data point belongs to.
    
    sil_score = silhouette_score(scaled_data_df, cluster_labels)
    
#silhouette_score() calculates the silhouette score, which measures how well each data point fits into its assigned cluster compared to other clusters.
  
    silhoutte_scores.append(sil_score)
    
#The silhouette score for the current value of k is appended to the silhouette_scores list. This allows  to keep track of how the silhouette score changes as you vary k.
 
    inertia.append(kmeans.inertia_)

#kmeans.inertia_ stores the inertia value, which measures how compact the clusters are. It's the sum of squared distances between data points and their assigned cluster centroids.

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title('KMeans Inertia for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.xticks(k_values)
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_values, silhoutte_scores, marker='o', color='orange')
plt.title('Silhouette Scores for Different Values of k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.xticks(k_values)
plt.grid(True)

plt.tight_layout()
plt.show()
```
